##Importar_Arquivo_Jason_GC

```python
# üì• Upload do JSON de credenciais
print("üîê Agora fa√ßa upload do arquivo .json de credenciais da conta de servi√ßo do Google Cloud:")
uploaded_json = files.upload()

# Captura o nome do arquivo JSON
json_cred = None
for fname in uploaded_json.keys():
    if fname.endswith(".json"):
        json_cred = fname

if not json_cred:
    raise Exception("‚ùå Nenhum arquivo .json de credenciais foi enviado!")

# Define a vari√°vel de ambiente para autentica√ß√£o
import os
os.environ["GOOGLE_APPLICATION_CREDENTIALS"] = json_cred

print(f"‚úÖ JSON de credenciais carregado: {json_cred}")
```

##Importar_Arquivo_Bucket

```python
from google.cloud import storage

# Configure o nome do bucket
bucket_name = "projeto-bi"
pasta_destino = "transacoes/"

# Conectando ao cliente GCS
client = storage.Client()

# Lista os arquivos dispon√≠veis
import os
arquivos_csv_no_colab = [f for f in os.listdir() if f.endswith('.csv')]

print("üìÇ Arquivos dispon√≠veis no ambiente:")
for i, nome in enumerate(arquivos_csv_no_colab):
    print(f"[{i}] {nome}")

# üëá Escolha os √≠ndices dos arquivos que deseja subir
# Exemplo: indices_escolhidos = [0, 2] para subir o 1¬∫ e 3¬∫ arquivos
indices_escolhidos = [0, 1]  # << ALTERE AQUI se quiser subir mais de um

# Faz upload para o bucket
bucket = client.bucket(bucket_name)
for i in indices_escolhidos:
    nome_arquivo = arquivos_csv_no_colab[i]
    blob = bucket.blob(pasta_destino + nome_arquivo)
    blob.upload_from_filename(nome_arquivo)
    print(f"‚úÖ {nome_arquivo} enviado para gs://{bucket_name}/{pasta_destino}{nome_arquivo}")
```
